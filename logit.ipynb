```python
import pandas as pd
```

```python
df = pd.read_csv("weight-height.csv").sample(100)
```

```python
print(df.shape)
print(df.head())
```

```python
df["Gender"] = df["Gender"].replace(["Male","Female"],[0,1])
```

```python
df.head(20)
```

```python
x = df[["Height"]]
y = df[["Gender"]]
```

```python
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x))
```

```python
from sklearn.linear_model import LogisticRegression
loc = LogisticRegression()
```

```python
loc.fit(x,y)
```

```python
loc.score(x,y)
```

```python
pred = loc.predict(x)
```

```python
loc.predict_proba(x)
```

```python
plt.scatter(x, y, color='blue', alpha=0.5, label='Data')
plt.plot(x , loc.predict_proba(x))
```

```python
import numpy as np
x_range = np.linspace(x.min(), x.max(), 500).reshape(-1,1)
y_pre = loc.predict_proba(x_range)[:,1]
```

```python
plt.scatter(x,y, color = "green", alpha=.6, label = "main Data")
plt.plot(x_range, y_pre, color= "b", linewidth = 2, label = "loc curve")
```

```python
from sklearn.metrics import confusion_matrix,precision_score, recall_score
cm = confusion_matrix(y, pred)
```

```python

sns.heatmap(cm, annot=True, cmap="rainbow")
```

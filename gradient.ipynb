# Linear Regression with gradient descent

```python
m = 0 #gradient
c = 0 # y intercept
l = 0.0001 #learning rate
n = int(len(x)) #number of datapoints
iterations = 1000

for i in range(iterations):
    pred = (m*x) + c
    der_m = (-2/n) *sum(x*(y-pred))
    der_c = (-2/n) *sum((y-pred))
    m = m-(l*der_m)
    c = c-(l*der_c)
    print(m,c)
```

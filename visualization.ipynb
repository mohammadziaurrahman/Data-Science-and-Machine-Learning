```python
df = pd.DataFrame({
    'x': [1, 5, 4, 8, 9, 6],
    'y': [10, 20, 30, 25, 40, 20],
    'class': ["a", "b", "c", "b", "c", "c"]
})

# Creating a pairplot with hue
sns.pairplot(df, hue='class')

# Show the plot
plt.show()
```

```python
df = pd.DataFrame({
    'x': [1, 5, 4, 8, 9, 6],
    'y': [10, 20, 30, 25, 40, 20],
    'class': ["a", "b", "c", "b", "c", "c"]
})
df
plt.figure(figsize=(4,3))
sns.violinplot(x='class', y='y', data=df, palette="Set2")
plt.xlabel("Class")
plt.ylabel("Value of y")
plt.title("Violin Plot")

plt.show()
```

```python
df = pd.DataFrame({
    'x': [1, 5, 4, 8, 9, 6],
    'y': [10, 20, 30, 25, 40, 20],
    'class': ["a", "b", "c", "b", "c", "c"]
})
df
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='y', fill=True, hue = "class")
plt.title('Density Distribution of Survived Passengers')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='y', hue="class", fill=True, palette="Set1")

# Updated Labels
plt.title('Density Distribution of Y Values by Class')
plt.xlabel('Y Values')
plt.ylabel('Density')

plt.show()
```
